```python
import numpy as np
import torch
from matplotlib import pyplot as plt
```

```python
train_loss_1 = np.load('result/train_loss_b1_basseline.npy')
train_acc_1 = np.load('result/train_acc_b1_basseline.npy')
test_loss_1 = np.load('result/test_loss_b1_basseline.npy')
test_acc_1 = np.load('result/test_acc_b1_basseline.npy')

train_loss_2 = np.load('result/train_loss_b1_da_at.npy')
train_acc_2 = np.load('result/train_acc_b1_da_at.npy')
test_loss_2 = np.load('result/test_loss_b1_da_at.npy')
test_acc_2 = np.load('result/test_acc_b1_da_at.npy')


train_loss_3 = np.load('result/train_loss_b1_da_at_P025.npy')
train_acc_3 = np.load('result/train_acc_b1_da_at_P025.npy')
test_loss_3 = np.load('result/test_loss_b1_da_at_P025.npy')
test_acc_3 = np.load('result/test_acc_b1_da_at_P025.npy')
```
